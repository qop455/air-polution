```python
import graphlab as gl
import datetime as dt
```

```python
def add_time_interval(ts, colname):
    lastcol = ts.column_names()[-1]
    ts.add_column(ts.shift(steps=1)[lastcol], name=colname)
    return ts
```

```python
target = 'PM2.5'
```

```python
date_feature = gl.SFrame.read_csv("../data/wind_level.csv")
date_feature['datetime'] = date_feature['datetime'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
ts = gl.TimeSeries(date_feature, index='datetime')
ts = ts[[target]]

#ts['dlt'] = ts['PM2.5'] - ts.shift(steps=1)['PM2.5']
ts.head()
```

```python
interval = 24
for i in range(interval):
    colname = 't-' + str(i+1)
    add_time_interval(ts, colname)
```

```python
ts = ts[ts.column_names()[::-1]]
```

```python
ts.remove_column(name=target)
```

```python
sf = ts.to_sframe()
df = sf.to_dataframe()
```

```python
df.to_csv('../data/' + target + '_timeseries.csv', index=False)
```
