```python
import pandas as pd

locations = {"二林":"EL", "斗六":"DL", "崙背":"LB", "麥寮":"ML", "臺西":"TS", "南投":"NT", "彰化":"CH", "線西":"SS"}
dataframe = pd.DataFrame()
```

```python
dataframe["datetime"] = pd.read_csv("../data/many/二林.csv").datetime
```

```python
for loc, abb in locations.iteritems():
    df = pd.read_csv("../data/many/" + loc + ".csv")
    df = df[[col for col in df.columns.tolist() if col not in ["datetime"]]]
    df.columns = [abb + col for col in df.columns.tolist()]
    dataframe = pd.concat([dataframe, df], axis=1)
```

```python
dataframe.to_csv("../data/many/all.csv", index=False)
```
