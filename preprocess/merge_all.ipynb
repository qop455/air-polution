```python
import pandas as pd

neighbors = {"崙背":"LB", "南投":"NT", "彰化":"CH"}
```

```python
target_cols = ["PM2.5", "AMB_TEMP", "RH", "RAINFALL", "WIND_SPEED", "WIND_DIREC"]
dataframe = pd.read_csv("../data/二林.csv", index_col="datetime")
dataframe = dataframe[target_cols]
dataframe = dataframe.rename(columns={"PM2.5":"PM25", "WIND_SPEED":"WS", "WIND_DIREC":"WD", "AMB_TEMP":"TEMP"})
dataframe.columns = ["EL" + "_" + col for col in dataframe.columns.tolist()]
```

```python
for loc, abb in neighbors.iteritems():
    df = pd.read_csv("../data/" + loc + ".csv", index_col="datetime")
    neighbor_cols = ["PM2.5", "WIND_SPEED", "WIND_DIREC"]
    df = df[neighbor_cols]
    df = df.rename(columns={"PM2.5":"PM25", "WIND_SPEED":"WS", "WIND_DIREC":"WD"})
    df.columns = [abb + "_" + col for col in df.columns.tolist()]
    dataframe = pd.concat([dataframe, df], axis=1)
```
