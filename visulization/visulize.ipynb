```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
```

```python
data = pd.read_csv("../data/Erlin.csv", parse_dates=True).fillna(0)
```

```python
data.info()
```

```python
targets = ['RAINFALL', 'RH', 'AMB_TEMP', 'WIND_SPEED', 'WIND_DIREC']
```

```python
data = data[targets]
```

```python
for target in targets:
    plt.rcParams['figure.figsize'] = [16.0, 8.0]
    plt.hist(data[target], bins=np.arange(0, max(data[target])), rwidth=0.8)
    plt.title(target)
    plt.xlabel('value')
    plt.ylabel('accumulation')
    fig = plt.gcf()
    plt.show()
```
