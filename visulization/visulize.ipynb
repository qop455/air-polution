```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
plt.rcParams['figure.figsize'] = [16.0, 8.0]
```

```python
data = pd.read_csv("../data/Erlin.csv", parse_dates=True).fillna(0)

targets = ['RAINFALL', 'RH', 'AMB_TEMP', 'WIND_SPEED', 'WIND_DIREC']

data = data[targets]
```

```python
for target in targets:
    plt.hist(data[target], bins=np.arange(0, max(data[target])), rwidth=0.8)
    plt.title(target)
    plt.xlabel('value')
    plt.ylabel('accumulation')
    fig = plt.gcf()
    plt.show()
```

```python
coef = pd.read_csv('../output/LR_coefficients.csv')

coef.describe()
```

```python
df_coefs = pd.DataFrame(dict(coef.mean()), index=[0])

df_coefs.describe()
```

```python
plt.bar(range(len(df_coefs.T.values)), df_coefs.T.values, width=0.3, color='C0', alpha=0.8)
plt.axhline(color='C0', xmin=0.01, xmax=0.99, alpha=0.8)
for x in range(len(df_coefs.columns)):
    if df_coefs.T.values[x] > 0.3 or df_coefs.T.values[x] < -0.3:
        plt.text(x - 0.15 , df_coefs.T.values[x] , str(round(df_coefs.T.values[x], 4)), color='k')
plt.xticks(range(len(df_coefs.T.values)), df_coefs.columns)
plt.xticks(rotation=90)
plt.title('LR Coefficients')
plt.xlabel('feature')
plt.ylabel('value')
plt.ylim([-6,1.5])
fig = plt.gcf()
plt.show()
```
