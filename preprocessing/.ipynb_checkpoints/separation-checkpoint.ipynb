```python
import pandas as pd
locations = ["EL", "DL", "LB", "ML", "TS", "NT", "CH", "SS"]
```

```python
inputfile = "../data/many/nan.csv"
outputfile = "../data/many/separated.csv"
```

```python
df = pd.read_csv(inputfile)
df.info()
```

```python
for index, row in df.iterrows():
    for loc in locations:
        ws = row[loc+"WIND_SPEED"]
        if ws >= 0 and ws <= 0.3:
            df.set_value(index, loc+'WIND_SPEED', 0)
        elif ws >= 0.3 and ws < 1.6:
            df.set_value(index, loc+'WIND_SPEED', 1)
        elif ws >= 1.6 and ws < 3.4:
            df.set_value(index, loc+'WIND_SPEED', 2)
        elif ws >= 3.4 and ws < 5.5:
            df.set_value(index, loc+'WIND_SPEED', 3)
        elif ws >= 5.5 and ws < 8.0:
            df.set_value(index, loc+'WIND_SPEED', 4)
        elif ws >= 8.0 and ws < 10.8:
            df.set_value(index, loc+'WIND_SPEED', 5)
        elif ws >= 10.8 and ws < 13.9:
            df.set_value(index, loc+'WIND_SPEED', 6)
        elif ws >= 13.9 and ws < 17.2:
            df.set_value(index, loc+'WIND_SPEED', 7)
        elif ws >= 17.2 and ws < 20.8:
            df.set_value(index, loc+'WIND_SPEED', 8)
        elif ws >= 20.8 and ws < 24.5:
            df.set_value(index, loc+'WIND_SPEED', 9)
        elif ws >= 24.5 and ws < 28.5:
            df.set_value(index, loc+'WIND_SPEED', 10)
        elif ws >= 28.5 and ws < 32.6:
            df.set_value(index, loc+'WIND_SPEED', 11)
        elif ws >= 32.6 and ws < 37.0:
            df.set_value(index, loc+'WIND_SPEED', 12)
        else:
            pass
            #print 'WIND SPEED ERROR:', ws, 'INDEX:', index
    
        ws = row[loc+"WIND_DIREC"]
        if ws >= 0 and ws < 22.5:
            df.set_value(index, loc+'WIND_DIREC', 0)
        elif ws >= 22.5 and ws < 67.5:
            df.set_value(index, loc+'WIND_DIREC', 45)
        elif ws >= 67.5 and ws < 112.5:
            df.set_value(index, loc+'WIND_DIREC', 90)
        elif ws >= 112.5 and ws < 157.5:
            df.set_value(index, loc+'WIND_DIREC', 135)
        elif ws >= 157.5 and ws < 202.5:
            df.set_value(index, loc+'WIND_DIREC', 180)
        elif ws >= 202.5 and ws < 247.5:
            df.set_value(index, loc+'WIND_DIREC', 225)
        elif ws >= 247.5 and ws < 292.5:
            df.set_value(index, loc+'WIND_DIREC', 270)
        elif ws >= 292.5 and ws < 337.5:
            df.set_value(index, loc+'WIND_DIREC', 315)
        elif ws >= 337.5 and ws <= 360:
            df.set_value(index, loc+'WIND_DIREC', 0)
        else:
            pass
            #print 'WIND DIREC ERROR:', ws, 'INDEX:', index
```
