```python
import pandas as pd
locations = ["EL", "DL", "LB", "ML", "TS", "NT", "CH", "SS"]
```

```python
df = pd.read_csv('../data/many/separated.csv')
df.head()
```

```python
for loc in locations:
    #rainfall = pd.get_dummies(df['RAINFALL']).rename(columns = lambda x: "RF_" + str(int(x)))
    direct = pd.get_dummies(df[loc+'WIND_DIREC']).rename(columns = lambda x: loc+"_WD_" + str(int(x/45)))
    #speed = pd.get_dummies(df[loc+'WIND_SPEED']).rename(columns = lambda x: loc+"_WS_" + str(int(x)))
    #rh = pd.get_dummies(df['RH']).rename(columns = lambda x: "RH_" + str(int(x)))
    df = pd.concat([df, direct], axis=1)
```

```python
drops = [loc+"WIND_DIREC" for loc in locations]
#drops = [inner for outer in drops for inner in outer]
```

```python
df.drop(drops, inplace=True, axis=1)
```
